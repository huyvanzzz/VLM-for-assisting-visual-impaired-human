```python
# ===== CELL 6: Compare Multiple Models (OPTIONAL) =====
configs = [
    "../configs/llava_config.yaml",
    "../configs/qwen_config.yaml",
]

all_results = {}
for config in configs:
    trainer = VLMTrainer(config)
    trainer.setup()
    trainer.train()
    results = trainer.evaluate()
    all_results[config] = results

# Plot comparison
import pandas as pd
df = pd.DataFrame(all_results).T
df.plot(kind='bar', title='Model Comparison')
```
